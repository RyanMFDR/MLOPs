## Input Data

```python
import pandas as pd

# Memuat data CSV
file_path = r"C:\Users\RyanMFDR\Documents\Project\MLOPs\data\JPTIIK.csv"
df = pd.read_csv(file_path)

# Menampilkan beberapa baris pertama untuk melihat struktur data
df.head()
```

## Drop Tabel tidak perlu & Duplikasi

```python
columns_to_drop = [
    "web-scraper-order",
    "web-scraper-start-url",
    "jurnal",
    "jurnal-href",
    "artikel-href",
    "artikel",
]
df.drop(columns=columns_to_drop, inplace=True, errors="ignore")

df = df.groupby("title", as_index=False).agg({"penulis": ", ".join})
df.head()
```

## Data Cleaning spaCy

```python
import spacy

nlp = spacy.load("en_core_web_sm")

def clean_text_spacy(text):
    text = text.lower()
    text = text.translate(str.maketrans("", "", string.punctuation))
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if not token.is_stop]
    return " ".join(tokens)

df["cleaned_title"] = df["title"].astype(str).apply(clean_text_spacy)

df.to_csv(r"C:\Users\RyanMFDR\Documents\Project\MLOPs\data\JPTIIK_cleaned.csv", index=False)

print(df[["title", "cleaned_title"]].head())
```
